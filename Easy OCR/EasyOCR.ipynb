```python
!pip install -q easyocr
import easyocr
import matplotlib.pyplot as plt
import cv2
```

```python
reader_obj_eng = easyocr.Reader(["en"])
reader_obj_fa  = easyocr.Reader(["fa"])

```

```python
result = reader_obj_eng.readtext("/content/text.jfif")
img1 = cv2.imread("/content/text.jfif")
img1 = cv2.cvtColor(img1 , cv2.COLOR_BGR2RGB)
plt.imshow(img1)
result
```

```python
for item in result :
  if item[2] > 0.5 :
    print(item)
```

#  Inference on a latin hand-writing text image

```python
Latin_HW = reader_obj_eng.readtext("/content/eng_handwriting.png")
Latin_HW
```

```python
img2 = cv2.imread("/content/eng_handwriting.png")
img2 = cv2.cvtColor(img2 , cv2.COLOR_BGR2RGB)
plt.imshow(img2)

item_list=[]
for item in Latin_HW :
  item_list.append(item[1])
print(item_list)
```
